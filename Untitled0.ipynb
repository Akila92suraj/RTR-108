\begin{equation}
cos(\sqrt{x})=\sum_{k=0}^\infty\frac{(-1)^k\cdot x^k} {(2\cdot k)!}=
=
\frac{(-1)^0\cdot x^0}{(2\cdot 0)!} +\frac{(-1)^1\cdot x^1}{(2\cdot (1)!}+\frac{(-1)^2\cdot x^2}{(2\cdot (2))!}+\frac{(-1)^3\cdot x^3}{(2\cdot (3))!}+\frac{(-1)^4\cdot x^4}{(2\cdot (4))!}
\end{equation}

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 50.0, 0.01)
s = np.cos(np.sqrt( t))

fig,ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About cos sqrt(t)')
ax.grid()

fig.savefig("test.png")
plt.show()
```
